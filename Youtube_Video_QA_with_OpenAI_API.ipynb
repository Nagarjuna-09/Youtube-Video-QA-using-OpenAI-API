```python
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.output_parsers import StrOutputParser
from langchain_openai import ChatOpenAI
import getpass
import os
from pytube import extract

# Extracting video id from url

def extract_video_id_split(url):
    video_id=extract.video_id(url)
    return video_id

print("Provide the video URL to analyze")
url = input()
video_id = extract_video_id_split(url)
# print(video_id)

# Extracts the transcript form a Youtube id (present in url)
from youtube_transcript_api import YouTubeTranscriptApi

input_data = YouTubeTranscriptApi.get_transcript(video_id)

paragraph = ' '.join([item['text'] for item in input_data])
# print(paragraph)
print("Learning from video complete...")

# Connects to open ai and provides result to the question
print("Provide OPEN AI API KEY")
os.environ["OPENAI_API_KEY"] = getpass.getpass()

model = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
parser = StrOutputParser()

while True:
  print("What do you want to know from the video?")
  question = input()

  messages = [
      SystemMessage(content=f"From the provided content {paragraph}, answer the questions. Strictly answer from the content provided. If the information is not provided in the content, respond as no information available"),
      HumanMessage(content=question),
  ]

  result = model.invoke(messages)
  parsed_result = parser.invoke(result)
  print(parsed_result)
  print("\n")
```
